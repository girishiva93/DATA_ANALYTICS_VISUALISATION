## Preprocess the kaggle dataset

```python
import pandas as pd

# region-to-locations mapping
on_mapping = region_mapping

# Convert it into location ‚Üí region mapping
location_to_region = {}
for region, locations in on_mapping.items():
    for loc in locations:
        location_to_region[loc.lower()] = region  # Use lowercase for safer matching

# Load weather data
weather_df = pd.read_csv("all_weather_data.csv")

# Add Region column using the mapping
weather_df['Region'] = weather_df['location'].str.lower().map(location_to_region)

# Extract Year and Month from date
weather_df['date'] = pd.to_datetime(weather_df['date'])
weather_df['Year'] = weather_df['date'].dt.year
weather_df['Month'] = weather_df['date'].dt.month

# Filter out rows with unmapped regions
weather_df = weather_df.dropna(subset=['Region'])

```

#### Utlise average to calculate the region temperature 

```python
# Calculate average monthly temps per region
avg_temp_by_region = weather_df.groupby(['Region', 'Year', 'Month']).agg({
    'min_temp ¬∞c': 'mean',
    'max_temp ¬∞c': 'mean'
}).reset_index()

# Rename columns
avg_temp_by_region.rename(columns={
    'min_temp ¬∞c': 'Avg_Min_Temp',
    'max_temp ¬∞c': 'Avg_Max_Temp'
}, inplace=True)
```

#### Preprocess Avian Dataset

```python
# Strip whitespace and normalize case
draft_df['Region'] = draft_df['Region'].str.strip().str.lower()
avg_temp_by_region['Region'] = avg_temp_by_region['Region'].str.strip().str.lower()
```

```python
draft_df['Year'] = draft_df['Year'].astype(int)
draft_df['Month'] = draft_df['Month'].astype(int)
avg_temp_by_region['Year'] = avg_temp_by_region['Year'].astype(int)
avg_temp_by_region['Month'] = avg_temp_by_region['Month'].astype(int)
```

```python
print("Regions in draft_df:", draft_df['Region'].unique())
print("Regions in avg_temp_by_region:", avg_temp_by_region['Region'].unique())

print("Years in draft_df:", draft_df['Year'].unique())
print("Years in avg_temp_by_region:", avg_temp_by_region['Year'].unique())

print("Months in draft_df:", draft_df['Month'].unique())
print("Months in avg_temp_by_region:", avg_temp_by_region['Month'].unique())
```

## Merge the average temperature into your draft

```python
merged_df = pd.merge(draft_df, avg_temp_by_region, on=['Region', 'Year', 'Month'], how='left')
```
